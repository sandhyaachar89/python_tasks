```python
data["category"]=data['Age'].apply(lambda x: "senior" if x > 50 else ("adult" if x >= 18 else "child"))
data
```

```python
data["category"] = pd.Categorical(data["category"])
print(data["category"].cat.categories)
print(data["category"].value_counts())
```

```python
data["Flight Status"] = pd.Categorical(data["Flight Status"])
print(data["Flight Status"].cat.categories)
print(data["Flight Status"].value_counts())
```

```python
data["Amount"] = data["Amount"].astype(float)
print(data["Amount"].dtypes)
print(data.iloc[:15,[15]])
```
