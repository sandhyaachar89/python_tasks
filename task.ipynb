```python
# create variables of different data types and print. their values and types and memory addresses/ids
name = "Anudeep" #string
age = 32 #integer
height = 5.9 #float
is_student = False #boolean
print("Name:", name, "Type:", type(name), "ID:", id(name))
print("Age:", age, "Type:", type(age), "ID:", id(age))
print("Height:", height, "Type:", type(height), "Height:", id(height) )
print("Is Student:", is_student, "Type:", type(is_student), "ID:", id(is_student) )
```
